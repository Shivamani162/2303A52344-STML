```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
file_path = '/content/drive/MyDrive/salary.xlsx.csv'
df = pd.read_csv(file_path)
print("Dataset preview:")
print(df.head())
features = df[['YearsExperience']]
target = df['Salary']
print("\nFeatures and Target Variable:")
print("Features:")
print(features.head())
print("Target:")
print(target.head())
plt.figure(figsize=(10, 6))
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.title('Years of Experience vs Salary')
plt.grid(True)
plt.show()
scaler_features = MinMaxScaler()
scaler_target = MinMaxScaler()
normalized_features = scaler_features.fit_transform(features)

df_normalized_features = pd.DataFrame(normalized_features, columns=['YearsExperience'])
print("\nNormalized Features:")
print(df_normalized_features.head())
normalized_target = scaler_target.fit_transform(target.values.reshape(-1, 1))
df_normalized_target = pd.DataFrame(normalized_target, columns=['Salary'])
print("\nNormalized Target:")
print(df_normalized_target.head())
```
